````python
def largest_and_smallest(lst):
    largest = lst[0]
    smallest = lst[0]

    for i in lst:
        if i > largest:
            largest = i
        if i < smallest:
            smallest = i

    return largest, smallest


lst = [1, 3, 5, 2, 6, 9, 4, 23]
print(largest_and_smallest(lst))
````

````python
# find the missing number in consecutive numbers in a set
def missing_number(set_1):
    n=len(set_1)
    total=(n+1)*(n+2)//2
    return abs(sum(set_1)-total)

set_1={1,2,3,5,6,7,8,9}
print("missing number is :",missing_number(set_1))
````

````python

## Find All Pairs in a Set with a Given Difference

### Problem Statement
"""
Given a **set of integers** and a number `k`, find all **pairs of elements** such that the **difference between them is equal to `k`**.

---

### Input

* A set of integers `set1`
* An integer `k`

---

### Output

* A set of pairs `(a, b)` such that:

```python
b - a = k
```

---

### Example

```python
set1 = {1, 5, 3, 4, 2}
k = 2
```

### Output

```python
{(1, 3), (3, 5), (2, 4)}
```

---

### Notes

* Each pair should satisfy `b = a + k`
* Avoid duplicate pairs
* Since input is a set, all elements are unique
* Order of pairs in output does not matter
"""
def all_pairs(set_1,k):
    pairs=set()
    for num in set_1:
        if num+k in set_1:
            pairs.add((num,num+k))
    return pairs
set_1={1,5,3,4,2}
print(all_pairs(set_1,2))
````
