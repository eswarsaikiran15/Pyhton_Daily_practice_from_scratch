```python
# Polymorphism
"""
polymorphism means Same method name → different behavior depending on object/type
"""

# method overriding
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

# usage
a1 = Dog()
a2 = Cat()

a1.sound()
a2.sound()
```

```python
#Operator Polymorphism

# integers
print(5 + 3)

# strings
print("Hello " + "World")

# lists
print([1, 2] + [3, 4])
```

```python
#Function Polymorphism (Duck Typing)
class Bird:
    def fly(self):
        print("Bird flies")

class Airplane:
    def fly(self):
        print("Airplane flies")

def start_flying(obj):
    obj.fly()

# usage
start_flying(Bird())
start_flying(Airplane())
```

```python
# Method Overloading (Python way)

def add(*args):
    return sum(args)

# usage
print(add(2, 3))
print(add(2, 3, 4))
print(add(1, 2, 3, 4))
```
