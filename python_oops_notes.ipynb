## 🔑 1. Introduction to OOP in Python

| Concept | Description |
|--------|-------------|
| **OOP Paradigm** | Programs are structured around **objects** (data + behavior) rather than procedures. |
| **Python's Nature** | Everything is an object (`int`, `list`, `function`, `module`). Supports multi-paradigm but excels in OOP. |
| **Class** | Blueprint/template for creating objects. Defines attributes & methods. |
| **Object** | Instance of a class. Holds state (data) and can perform actions (methods). |

### 🧱 Basic Syntax
```python
class Dog:
    # Class attribute (shared by all instances)
    species = "Canine"

    def __init__(self, name, age):
        # Instance attributes (unique per object)
        self.name = name
        self.age = age

    def bark(self):
        return f"{self.name} says Woof!"

# Instantiation
dog1 = Dog("Buddy", 3)
print(dog1.bark())  # Buddy says Woof!
```

## 🏛️ 2. The 4 Pillars of OOP

### 🔒 2.1 Encapsulation
**Theory:** Bundling data and methods that operate on that data, while restricting direct access to some components.

**Python Implementation:** Python uses **naming conventions**, not strict access modifiers.
| Prefix | Meaning | Access |
|--------|---------|--------|
| `name` | Public | Accessible anywhere |
| `_name` | Protected (convention) | Internal use, but accessible |
| `__name` | Private (name-mangled) | `_ClassName__name` internally |

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # "private"

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def get_balance(self):
        return self.__balance

acc = BankAccount("Alice", 1000)
acc.deposit(500)
print(acc.get_balance())  # 1500
# print(acc.__balance)   # AttributeError
# print(acc._BankAccount__balance)  # Works, but violates encapsulation
```

> 💡 **Interview Tip:** Python doesn't enforce access control. It relies on developer discipline (`_` and `__` are hints, not barriers).

### 🎭 2.2 Abstraction
**Theory:** Hiding complex implementation details and exposing only essential features.

**Python Implementation:** Use the `abc` (Abstract Base Classes) module.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14159 * self.radius

# shape = Shape()  # TypeError: Can't instantiate abstract class
c = Circle(5)
print(c.area())  # 78.53975
```

> 💡 **Interview Tip:** Abstract classes define interfaces. Concrete classes implement them. Prevents incomplete subclass implementations.

### 🔗 2.3 Inheritance
**Theory:** Mechanism where a new class derives attributes & methods from an existing class. Promotes code reuse.

#### Types:
- **Single:** `B(A)`
- **Multiple:** `C(A, B)`
- **Multilevel:** `C → B → A`
- **Hierarchical:** `B(A)`, `C(A)`
- **Hybrid:** Combination

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        return f"{self.brand} engine started"

class Car(Vehicle):
    def __init__(self, brand, doors):
        super().__init__(brand)  # Call parent constructor
        self.doors = doors

    def start(self):
        return f"{super().start()} with {self.doors} doors"

car = Car("Tesla", 4)
print(car.start())  # Tesla engine started with 4 doors
```

#### 🔍 Method Resolution Order (MRO)
Python uses **C3 Linearization** to resolve multiple inheritance.
```python
class A: pass
class B(A): pass
class C(A): pass
class D(B, C): pass

print(D.__mro__)  # (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
```
> 💡 **Interview Tip:** `super()` follows MRO. Diamond problem is solved by C3 linearization, not by choosing left/right arbitrarily.

### 🔄 2.4 Polymorphism
**Theory:** Same interface, different underlying forms. "Many shapes".

#### Forms in Python:
1. **Method Overriding** (Runtime polymorphism)
2. **Duck Typing** ("If it walks like a duck...")
3. **Operator Overloading** (Magic methods)

```python
# Duck Typing
def make_sound(animal):
    animal.speak()  # Works if object has .speak()

class Cat:
    def speak(self): print("Meow")

class Dog:
    def speak(self): print("Woof")

make_sound(Cat())
make_sound(Dog())
```

```python
# Operator Overloading
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)  # Vector(4, 6)
```

> 💡 **Interview Tip:** Python doesn't support method overloading natively. Use `*args`, `**kwargs`, `typing.overload`, or default parameters.

## 🛠️ 3. Methods & Decorators

| Type | Syntax | Bound To | Use Case |
|------|--------|----------|----------|
| Instance | `def method(self):` | Object instance | Access/modify instance state |
| Class | `@classmethod def method(cls):` | Class | Alternative constructors, class-level operations |
| Static | `@staticmethod def method():` | None | Utility functions related to class |

```python
class MathUtils:
    factor = 10

    def __init__(self, value):
        self.value = value

    @classmethod
    def from_string(cls, s: str):
        return cls(int(s))

    @staticmethod
    def add(a, b):
        return a + b

    def scale(self):
        return self.value * self.factor

obj = MathUtils.from_string("5")
print(obj.scale())      # 50
print(MathUtils.add(2, 3))  # 5
```

## 🪄 4. Magic / Dunder Methods

| Method | Purpose |
|--------|---------|
| `__init__` | Constructor |
| `__new__` | Object creation (before `__init__`) |
| `__str__` | User-friendly string |
| `__repr__` | Developer-friendly string |
| `__len__` | `len(obj)` |
| `__getitem__`, `__setitem__` | Indexing `obj[i]` |
| `__call__` | Make object callable `obj()` |
| `__enter__`, `__exit__` | Context managers (`with`) |
| `__eq__`, `__lt__`, etc. | Comparison operators |

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} yrs"

    def __repr__(self):
        return f"Person('{self.name}', {self.age})"

    def __eq__(self, other):
        return isinstance(other, Person) and self.age == other.age

p = Person("Alex", 30)
print(p)        # Alex, 30 yrs
print(repr(p))  # Person('Alex', 30)
```

> 💡 **Interview Tip:** Always implement `__repr__`. `__str__` is optional. If only one is defined, Python falls back to it.

## 🔧 5. Properties & Controlled Access

Pythonic way to replace Java-style getters/setters.

```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            raise ValueError("Below absolute zero!")
        self._celsius = value

    @property
    def fahrenheit(self):
        return self._celsius * 9/5 + 32

t = Temperature(25)
t.celsius = 30  # Triggers setter
print(t.fahrenheit)  # 86.0
```

## 🧠 6. Advanced OOP Concepts (Interview Heavy)

### 🔹 `__new__` vs `__init__`
| `__new__` | `__init__` |
|-----------|------------|
| Creates instance | Initializes instance |
| Static method (implicitly) | Instance method |
| Returns object | Returns `None` |
| Used in metaclasses, immutable types | Used for 99% of classes |

```python
class Singleton:
    _instance = None

    def __new__(cls, *args, **kwargs):
        if not cls._instance:
            cls._instance = super().__new__(cls)
        return cls._instance
```

### 🔹 `__slots__` (Memory Optimization)
Prevents dynamic attribute creation, saves memory.
```python
class Point:
    __slots__ = ['x', 'y']  # No __dict__ created
```

## 🎤 7. Top 20 Interview Questions + Answers

1. **What is `self` in Python?**
   Reference to the instance. Explicit first parameter in instance methods. Not a keyword.

2. **Difference between `@classmethod` and `@staticmethod`?**
   `@classmethod` receives `cls` (can access/modify class state). `@staticmethod` receives neither `self` nor `cls` (utility function).

3. **How to achieve method overloading in Python?**
   Use default args, `*args`/`**kwargs`, or `typing.overload` decorator. Python doesn't support true overloading.

4. **What is MRO? How is it calculated?**
   Method Resolution Order. Uses C3 linearization. View with `ClassName.__mro__` or `ClassName.mro()`.

5. **Explain `super()` with multiple inheritance.**
   Calls next class in MRO chain. Ensures cooperative inheritance, avoids diamond problem.

6. **`__str__` vs `__repr__`?**
   `__str__` for users, `__repr__` for developers. `repr()` should ideally recreate object.

7. **How is data hiding achieved?**
   Naming conventions: `_protected`, `__private` (name-mangled). No strict enforcement.

8. **What is duck typing? Give example.**
   Type determined by behavior, not inheritance. `if hasattr(obj, 'quack'):` or just call method directly.

9. **When to use `__new__`?**
   When subclassing immutable types (`int`, `str`, `tuple`), or implementing singletons/factories.

10. **Composition vs Inheritance?**
    Inheritance = "is-a", Composition = "has-a". Prefer composition for flexibility & loose coupling.

11. **What are Abstract Base Classes?**
    Define interfaces using `abc.ABC` & `@abstractmethod`. Prevent incomplete subclasses.

12. **Can you access private variables?**
    Yes, via `_ClassName__var`. But it breaks encapsulation & is discouraged.

13. **What is `__slots__`?**
    Restricts instance attributes, saves memory, prevents `__dict__` creation.

14. **How to make an object callable?**
    Implement `__call__` method.

15. **What is a metaclass?**
    "Class of a class". Controls class creation. Default is `type`. Advanced topic.

16. **Explain Liskov Substitution Principle.**
    Subclasses should be usable anywhere base class is expected without breaking behavior.

17. **How does Python handle multiple inheritance conflicts?**
    C3 linearization ensures deterministic MRO. Left-to-right, depth-first, but respects inheritance order.

18. **What is `__init_subclass__`?**
    Hook called when a class is subclassed. Useful for plugin registration or validation.

19. **Difference between old-style & new-style classes?**
    Python 2: old-style didn't inherit `object`. Python 3: all classes inherit `object` automatically.

20. **How to implement a context manager in OOP?**
    Implement `__enter__` and `__exit__`, or use `contextlib.contextmanager` decorator.

1. **Favor composition over inheritance** unless there's a true "is-a" relationship.
2. **Use properties** instead of explicit getters/setters.

## 🔒 1. ENCAPSULATION

### 📖 Definition
Encapsulation bundles **data (attributes)** and **behavior (methods)** that operate on that data within a single unit (class), while **restricting direct external access** to internal state. It protects object integrity and enforces controlled interaction.

### 🐍 Python's Implementation
Python **does not enforce** strict access control. It relies on **naming conventions** and **developer discipline**.

| Convention | Meaning | Access Level | Name Mangling? |
|------------|---------|--------------|----------------|
| `name` | Public | Anywhere | ❌ |
| `_name` | Protected (convention) | Internal/module | ❌ |
| `__name` | Private (convention) | Class-only | ✅ (`_ClassName__name`) |

### 🧩 Types & Patterns

#### 🔹 1. Basic Encapsulation (Conventions)
```python
class User:
    def __init__(self, username, password):
        self.username = username
        self._password_hash = self._hash(password)  # "protected"
        self.__api_key = "secret_123"              # "private" (mangled)

    def _hash(self, pwd: str) -> str:
        return f"hashed_{pwd}"

    def verify(self, pwd: str) -> bool:
        return self._hash(pwd) == self._password_hash
```

#### 🔹 2. Pythonic Encapsulation (`@property`)
Replaces Java-style getters/setters. Adds validation & computed attributes.
```python
class BankAccount:
    def __init__(self, balance: float):
        self._balance = balance  # internal state

    @property
    def balance(self) -> float:
        return self._balance

    @balance.setter
    def balance(self, amount: float):
        if amount < 0:
            raise ValueError("Balance cannot be negative")
        self._balance = amount

    @property
    def is_overdrawn(self) -> bool:
        return self._balance < 0  # computed/read-only
```

#### 🔹 3. Read-Only & Write-Only Attributes
```python
class Sensor:
    def __init__(self, value: float):
        self._value = value

    @property
    def value(self) -> float:  # Read-only
        return self._value

class Logger:
    _log = []

    @property
    def log(self):  # Write-only (raises error on read)
        raise AttributeError("Log is write-only")

    @log.setter
    def log(self, entry: str):
        self._log.append(entry)
```

### 🎯 Advanced/Interview Concepts
| Concept | Explanation |
|---------|-------------|
| **Name Mangling** | `__var` → `_ClassName__var`. Prevents accidental override in subclasses, not true privacy. |
| `__slots__` | Restricts dynamic attribute creation, enforces encapsulation & saves memory. |
| **Validation Layer** | Use `@property` setters to enforce invariants (type, range, format). |
| **Immutable Objects** | Combine `__slots__` + no setters + `@dataclass(frozen=True)` for strict encapsulation. |

### ✅ Best Practices
- ✅ Use `@property` instead of explicit `get_x()`/`set_x()`
- ✅ Prefix internal APIs with `_` to signal "not part of public contract"
- ✅ Use `__` only when subclass collision is likely (e.g., framework base classes)
- ✅ Document public API clearly; treat `_` as internal implementation details
- ❌ Never rely on `__` for security; it's easily bypassed (`obj._Class__var`)

## 🎭 2. ABSTRACTION

### 📖 Definition
Abstraction hides **complex implementation details** and exposes only **essential features/interfaces**. Users interact with simplified contracts without knowing internal mechanics.

### 🐍 Python's Implementation
Python supports two abstraction paradigms:
1. **Nominal Subtyping** (`abc.ABC`) → "You must inherit from me"
2. **Structural Subtyping** (`typing.Protocol`) → "You must behave like me"

### 🧩 Types & Patterns

#### 🔹 1. Abstract Base Classes (`abc`)
Enforces method implementation at instantiation time.
```python
from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    @abstractmethod
    def process_payment(self, amount: float) -> bool:
        pass

    @abstractmethod
    def refund(self, transaction_id: str) -> bool:
        pass

    # Concrete method (shared logic)
    def log_transaction(self, status: str):
        print(f"[LOG] {status}")

class StripeGateway(PaymentGateway):
    def process_payment(self, amount: float) -> bool:
        self.log_transaction("Processing Stripe")
        return True

    def refund(self, transaction_id: str) -> bool:
        return True

# pg = PaymentGateway()  # ❌ TypeError: Can't instantiate abstract class
sg = StripeGateway()
print(sg.process_payment(100.0))  # ✅ Works
```

#### 🔹 2. Structural Abstraction (`typing.Protocol`)
Modern Python alternative. No inheritance required. Based on duck typing.
```python
from typing import Protocol

class Serializable(Protocol):
    def to_dict(self) -> dict: ...
    def from_dict(self, data: dict) -> None: ...

class User:
    def to_dict(self) -> dict:
        return {"id": 1, "name": "Alice"}

    def from_dict(self, data: dict) -> None:
        self.id = data["id"]
        self.name = data["name"]

def save(entity: Serializable) -> None:
    data = entity.to_dict()
    print(f"Saving: {data}")

save(User())  # ✅ Passes static check & runtime
```

#### 🔹 3. Partial vs Complete Abstraction
```python
class TemplateProcessor(ABC):
    @abstractmethod
    def load_template(self) -> str: ...

    def render(self) -> str:
        tmpl = self.load_template()
        return f"<html>{tmpl}</html>"  # Concrete uses abstract

class HTMLProcessor(TemplateProcessor):
    def load_template(self) -> str:
        return "<body>Hello</body>"
```

### 🎯 Advanced/Interview Concepts
| Concept | Explanation |
|---------|-------------|
| `ABC` vs `Protocol` | `ABC` enforces inheritance; `Protocol` enforces shape/behavior. Prefer `Protocol` for flexibility. |
| `@abstractmethod` on `@property` | Valid: `@property` + `@abstractmethod` enforces getter/setter in subclasses. |
| `__subclasshook__` | Override to customize `issubclass()` behavior for virtual subclasses. |
| **Interface Segregation** | Keep abstract classes small. Split fat interfaces into focused `Protocol`s. |

### ✅ Best Practices
- ✅ Use `Protocol` for duck-typed interfaces (modern Python standard)
- ✅ Use `ABC` when you need shared base implementation or strict inheritance enforcement
- ✅ Keep abstract methods minimal; provide concrete helpers where possible
- ✅ Document expected signatures & side effects in abstract contracts
- ❌ Avoid deep ABC hierarchies; prefer composition + small protocols

## 🔄 3. POLYMORPHISM

### 📖 Definition
Polymorphism ("many forms") allows objects of different types to be treated uniformly through a **common interface**. The same method call behaves differently based on the object's actual class at runtime.

### 🐍 Python's Implementation
Python achieves polymorphism through:
1. **Dynamic Dispatch** (runtime method resolution)
2. **Duck Typing** (behavior over inheritance)
3. **Operator Overloading** (dunder methods)
4. **Single/Multiple Dispatch** (`functools`)

### 🧩 Types & Patterns

#### 🔹 1. Duck Typing (Runtime Polymorphism)
"If it walks like a duck and quacks like a duck, it's a duck."
```python
class Cat:
    def speak(self) -> str: return "Meow"

class Robot:
    def speak(self) -> str: return "Beep Boop"

def announce(entity) -> None:
    print(entity.speak())  # No type check, just behavior

announce(Cat())   # Meow
announce(Robot()) # Beep Boop
```

#### 🔹 2. Method Overriding (Inheritance-based)
```python
class Shape:
    def area(self) -> float:
        raise NotImplementedError

class Circle(Shape):
    def __init__(self, r: float): self.r = r
    def area(self) -> float: return 3.14159 * self.r ** 2

class Square(Shape):
    def __init__(self, side: float): self.side = side
    def area(self) -> float: return self.side ** 2

def print_area(s: Shape) -> None:
    print(f"Area: {s.area():.2f}")

print_area(Circle(5))   # Area: 78.54
print_area(Square(4))   # Area: 16.00
```

#### 🔹 3. Operator Overloading (Dunder Methods)
```python
from typing import Self

class Vector:
    def __init__(self, x: float, y: float):
        self.x, self.y = x, y

    def __add__(self, other: Self) -> Self:
        return Vector(self.x + other.x, self.y + other.y)

    def __mul__(self, scalar: float) -> Self:
        return Vector(self.x * scalar, self.y * scalar)

    def __repr__(self) -> str:
        return f"Vector({self.x}, {self.y})"

v1, v2 = Vector(1, 2), Vector(3, 4)
print(v1 + v2)       # Vector(4, 6)
print(v1 * 3)        # Vector(3, 6)
```

#### 🔹 4. Single Dispatch (`functools`)
Type-based function overloading at runtime.
```python
from functools import singledispatch

@singledispatch
def process(data):
    raise NotImplementedError("Unsupported type")

@process.register
def _(data: str):
    return f"Processing string: {data.upper()}"

@process.register
def _(data: int):
    return f"Processing number: {data * 2}"

@process.register(list)
def _(data: list):
    return f"Processing list with {len(data)} items"

print(process("hello"))  # Processing string: HELLO
print(process(42))       # Processing number: 84
print(process([1,2,3]))  # Processing list with 3 items
```

### 🎯 Advanced/Interview Concepts
| Concept | Explanation |
|---------|-------------|
| **Dynamic Dispatch** | Python resolves methods at runtime via `__mro__`. Enables duck typing. |
| `@singledispatchmethod` | Class method version of `singledispatch` (Python 3.8+). |
| **No Compile-Time Overloading** | Python ignores duplicate method names; last definition wins. Use defaults/`*args`/`singledispatch`. |
| **Liskov Substitution Principle** | Subclasses must not break polymorphic expectations of base class. |
| **Performance** | Dynamic dispatch has slight overhead. Cache hot paths or use `__slots__` if critical. |

### ✅ Best Practices
- ✅ Favor duck typing + `Protocol` over `isinstance()` checks
- ✅ Implement dunder methods consistently (`__eq__` → `__hash__`, `__lt__` → `__le__`)
- ✅ Use `@singledispatch` for clean type-based routing instead of `if/elif isinstance`
- ✅ Keep polymorphic interfaces small & focused (ISP)
- ❌ Avoid `type(obj) == SomeClass`; use `isinstance()` or duck typing
- ❌ Don't overload methods by redefining in same class; use defaults or dispatch

## 🔑 1. What is Inheritance?

**Definition:** Inheritance is an OOP mechanism where a new class (**child/derived/subclass**) acquires attributes and methods from an existing class (**parent/base/superclass**).

**Why Use It?**
- ✅ **Code Reusability:** Write once, reuse across related classes
- ✅ **Extensibility:** Add/modify behavior without altering base code
- ✅ **Polymorphism Foundation:** Enables runtime method dispatch
- ✅ **Logical Hierarchy:** Models real-world "is-a" relationships

**Basic Syntax:**
```python
class Parent:
    pass

class Child(Parent):  # Child inherits from Parent
    pass
```

---

## 🧱 2. How Python Implements Inheritance

| Feature | Python Behavior |
|---------|----------------|
| **Root Class** | All classes implicitly inherit from `object` (Python 3) |
| **Dynamic Lookup** | Attribute search follows MRO at runtime, not compile-time |
| **No Access Modifiers** | Uses `_` and `__` conventions, not enforcement |
| **Multiple Inheritance** | Fully supported (unlike Java/C#) |
| **Inspection** | `Class.__bases__`, `Class.__mro__`, `issubclass()`, `isinstance()` |

## 🔗 3. Types of Inheritance (Definitions + Examples)

### 🟦 3.1 Single Inheritance
**Definition:** A subclass inherits from exactly one parent class.

**Structure:** `Child → Parent`

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Dog(Animal):  # Single inheritance
    def speak(self):
        return f"{self.name} says Woof!"

dog = Dog("Buddy")
print(dog.speak())  # Buddy says Woof!
print(issubclass(Dog, Animal))  # True
```
> 💡 **When to Use:** Clear "is-a" relationship, simple hierarchies, no ambiguity in method resolution.

---

### 🟩 3.2 Multiple Inheritance
**Definition:** A subclass inherits from two or more parent classes.

**Structure:** `Child → Parent1, Parent2, ...`

```python
class Flyer:
    def fly(self):
        return "Flying high ✈️"

class Swimmer:
    def swim(self):
        return "Swimming fast 🏊"

class Duck(Flyer, Swimmer):  # Multiple inheritance
    def quack(self):
        return "Quack! 🦆"

duck = Duck()
print(duck.fly())   # Flying high ✈️
print(duck.swim())  # Swimming fast 🏊
print(duck.quack()) # Quack! 🦆
```

**⚠️ Critical Rule:** Always use `super()` for cooperative initialization to avoid missed parent constructors or duplicate calls.

```python
class A:
    def __init__(self):
        print("A init")
        super().__init__()

class B:
    def __init__(self):
        print("B init")
        super().__init__()

class C(A, B):  # MRO: C → A → B → object
    def __init__(self):
        print("C init")
        super().__init__()

C()
# Output:
# C init
# A init
# B init
```
> 💡 **Interview Tip:** Python resolves conflicts using **C3 Linearization (MRO)**, not left-to-right DFS. Order in `class C(A, B):` matters.

---

### 🟨 3.3 Multilevel Inheritance
**Definition:** A chain of inheritance where a class inherits from a derived class, forming multiple levels.

**Structure:** `GrandChild → Child → Parent`

```python
class Vehicle:
    def start_engine(self):
        return "Engine started 🔑"

class Car(Vehicle):
    def honk(self):
        return "Beep beep 🚗"

class ElectricCar(Car):
    def charge(self):
        return "Charging battery 🔋"

ev = ElectricCar()
print(ev.start_engine())  # From Vehicle
print(ev.honk())          # From Car
print(ev.charge())        # From ElectricCar
```
> 💡 **When to Use:** Progressive specialization (e.g., `Model → Version → Trim`). Avoid chains >3 levels deep.

---

### 🟧 3.4 Hierarchical Inheritance
**Definition:** Multiple subclasses inherit from a single base class.

**Structure:** `Child1 → Parent ← Child2`

```python
class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    def work(self):
        return f"{self.name} is working 💼"

class Developer(Employee):
    def code(self):
        return f"{self.name} writes Python 🐍"

class Designer(Employee):
    def design(self):
        return f"{self.name} creates UI/UX 🎨"

dev = Developer("Alice", 101)
des = Designer("Bob", 102)

print(dev.work())   # Alice is working 💼
print(dev.code())   # Alice writes Python 🐍
print(des.design()) # Bob creates UI/UX 🎨
```
> 💡 **When to Use:** Shared base behavior with divergent specializations (e.g., payment methods, notification types).

---

### 🟥 3.5 Hybrid Inheritance
**Definition:** A combination of two or more inheritance types (usually multiple + multilevel).

**Structure:** Complex DAG (Directed Acyclic Graph)

```python
class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, name, emp_id):
        super().__init__(name)
        self.emp_id = emp_id

class Freelancer:
    def __init__(self, rate):
        self.rate = rate

class TechLead(Employee, Freelancer):  # Hybrid: Multilevel + Multiple
    def __init__(self, name, emp_id, rate, team):
        Employee.__init__(self, name, emp_id)
        Freelancer.__init__(self, rate)
        self.team = team

    def manage(self):
        return f"{self.name} leads team of {self.team}"

lead = TechLead("Charlie", 201, 150, 5)
print(lead.manage())          # Charlie leads team of 5
print(lead.name, lead.rate)   # Charlie 150
```
> 💡 **Interview Tip:** Hybrid inheritance is powerful but error-prone. Use `super()` consistently or prefer composition if hierarchy becomes tangled.

## ⚙️ 4. Core Mechanisms: `super()` & MRO

### 🔹 `super()` Function
- Returns a proxy object that delegates method calls to the **next class in MRO**
- Enables **cooperative inheritance** (all parent `__init__`/methods run in order)
- Syntax: `super()` (Python 3) or `super(ChildClass, self)` (Python 2)

```python
class A:
    def show(self):
        print("A")
        super().show()  # Calls next in MRO

class B(A):
    def show(self):
        print("B")
        super().show()

class C(A):
    def show(self):
        print("C")
        super().show()

class D(B, C):
    def show(self):
        print("D")
        super().show()

D().show()
# Output: D → B → C → A → (object has no show, stops safely if handled)
```

### 🔹 Method Resolution Order (MRO) & C3 Linearization
Python uses **C3 Linearization** to determine method lookup order. It guarantees:
1. **Consistency:** Subclass appears before parent
2. **Left-to-Right Order:** Preserves declaration order
3. **Monotonicity:** No class appears twice

```python
class A: pass
class B(A): pass
class C(A): pass
class D(B, C): pass

print(D.__mro__)
# (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)

# Alternative inspection
print(D.mro())
```

### 🔍 The Diamond Problem
```python
class A:
    def process(self): print("A.process")

class B(A):
    def process(self):
        super().process()
        print("B.process")

class C(A):
    def process(self):
        super().process()
        print("C.process")

class D(B, C):
    def process(self):
        super().process()
        print("D.process")

D().process()
# Output: A.process → C.process → B.process → D.process
```
✅ **C3 ensures `A.process()` runs exactly once**, unlike traditional DFS which would call it twice.

## 💡 5. Interview-Heavy Concepts

| Concept | Explanation | Python Snippet |
|---------|-------------|----------------|
| **Method Overriding** | Child replaces parent method | `def speak(self): ...` |
| **Method Extension** | Child adds to parent behavior | `super().work(); print("Extra")` |
| `issubclass()` | Checks class inheritance | `issubclass(Dog, Animal)` → `True` |
| `isinstance()` | Checks object inheritance | `isinstance(dog, Animal)` → `True` |
| **`__bases__`** | Tuple of direct parents | `Dog.__bases__` → `(<class 'Animal'>,)` |
| **Liskov Substitution** | Subclass must behave like base | `def process(x: Animal): x.speak()` |

## ✅ 6. Best Practices & Pitfalls

### 🟢 Do's
- ✅ Use `super()` for **cooperative multiple inheritance**
- ✅ Keep inheritance depth ≤ 3 levels
- ✅ Prefer **composition** over deep inheritance (`has-a` vs `is-a`)
- ✅ Use `abc.ABC` for interface-like base classes
- ✅ Document MRO expectations in complex hierarchies

### 🔴 Don'ts
- ❌ Call parent constructors directly (`Parent.__init__(self)`) in multiple inheritance
- ❌ Mix instance/class/static methods inconsistently across hierarchy
- ❌ Create circular inheritance (`A(B)` and `B(A)`)
- ❌ Override methods without calling `super()` when parent state matters
- ❌ Ignore `__mro__` when debugging method conflicts

```python
# Inspect hierarchy
print(MyClass.__bases__)      # Direct parents
print(MyClass.__mro__)        # Full resolution order
print(MyClass.mro())          # List version

# Check relationships
print(issubclass(Child, Parent))
print(isinstance(obj, Parent))

# View source/origin of a method
import inspect
print(inspect.getmembers(MyClass, predicate=inspect.isfunction))
```
