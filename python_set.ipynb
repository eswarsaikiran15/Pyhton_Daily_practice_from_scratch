# Sets in Python

## Definition

A **set** in Python is an **unordered, mutable collection of unique elements**.

### Key Properties

* No duplicates → automatically removes repeated values
* Unordered → no indexing (`set[0]` ❌)
* Mutable → elements can be added or removed
* Elements must be **hashable** (e.g., `int`, `str`, `tuple`)

### Example

```python
s = {1, 2, 3, 3}
print(s)   # Output: {1, 2, 3}
```

---

## Types of Sets

### 1. Set (Mutable)

```python
s = {1, 2, 3}
```

### 2. Frozen Set (Immutable)

```python
fs = frozenset([1, 2, 3])
# fs.add(4) ❌ Not allowed
```

---

## Built-in Functions for Sets

* `len(s)` → Number of elements
* `max(s)` → Largest element
* `min(s)` → Smallest element
* `sum(s)` → Sum of elements
* `sorted(s)` → Returns a sorted **list**
* `set(iterable)` → Converts iterable to set

### Example

```python
s = {3, 1, 2}
print(sorted(s))  # [1, 2, 3]
```

---

## Set Methods

### 1. Adding Elements

* `add(x)` → Adds a single element
* `update(iterable)` → Adds multiple elements

```python
s = {1, 2}
s.add(3)
s.update([4, 5])
```

---

### 2. Removing Elements

* `remove(x)` → Raises error if element not found
* `discard(x)` → No error if element not found
* `pop()` → Removes a random element
* `clear()` → Removes all elements

```python
s.remove(2)
s.discard(10)  # Safe
```

---

### 3. Set Operations

#### Union (All elements)

* `s1.union(s2)` or `s1 | s2`

#### Intersection (Common elements)

* `s1.intersection(s2)` or `s1 & s2`

#### Difference (Only in s1)

* `s1.difference(s2)` or `s1 - s2`

#### Symmetric Difference (Non-common elements)

* `s1.symmetric_difference(s2)` or `s1 ^ s2`

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)  # {1, 2, 3, 4, 5}
print(a & b)  # {3}
print(a - b)  # {1, 2}
print(a ^ b)  # {1, 2, 4, 5}
```

---

### 4. Set Relationship Methods

* `issubset()` → Checks if a set is part of another
* `issuperset()` → Checks if a set contains another
* `isdisjoint()` → Checks if sets have no common elements

```python
a = {1, 2}
b = {1, 2, 3}

print(a.issubset(b))   # True
print(b.issuperset(a)) # True
```

---

### 5. Copying

* `copy()` → Creates a shallow copy of the set

---

## When to Use Sets

### Use Sets When:

* Removing duplicates
* Fast membership testing (`O(1)` complexity)
* Performing union, intersection, and difference operations

### Avoid Sets When:

* Order matters
* Indexing is required
* Duplicate counts are needed → use `list` or `collections.Counter`

---

## Key Insight

Understanding **when to use sets** is more important than memorizing methods.
Sets are powerful for **data cleaning, filtering, and comparison tasks**, especially in real-world data processing and analysis.

## Most Used Functions and Methods on Sets

| Function / Method         | Equivalent Operator | Explanation                                 |
| ------------------------- | ------------------- | ------------------------------------------- |
| `set.add(value)`          | —                   | Adds the specified element into a set       |
| `set1.union(set2)`        | `set1 \| set2`      | Merges two given sets                       |
| `set1.intersection(set2)` | `set1 & set2`       | Finds common elements of both sets          |
| `set1.difference(set2)`   | `set1 - set2`       | Finds elements in set1 that are not in set2 |
| `set.clear()`             | —                   | Empties the set                             |
| `set()`                   | —                   | Creates a set or converts iterable to a set |
| `set1.isdisjoint(set2)`   | —                   | Checks if both sets have no common elements |


````python
set_1={1,1,2,2,3,3,4,4}
set_2={3,4,5}
print("set 1 is",set_1)
print("set 2 is",set_2)

set_1.add("kiran")
print("set 1 is",set_1)
# set_1.add(["kiran"])
# print("set 1 is",set_1) # this gets an error called unhashable type: 'list'
set_1.add((3,4))
print("set 1 is",set_1) # tuple is added and string is also added in set
set_1.remove((3,4))
set_2.remove(3)
print("set 1 is",set_1)
print("set 2 is",set_2)
````
